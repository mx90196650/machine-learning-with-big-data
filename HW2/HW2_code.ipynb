```python
import pandas as pd 
import numpy as np
import sklearn
from sklearn import tree
from sklearn import naive_bayes
from sklearn import model_selection
from sklearn import metrics
```

```python
df = pd.read_csv("house-votes-84.data", header=None, na_values="?")
df_numeric = df.replace(["republican", "democrat", "y", "n"], [1, 0, 1, 0])
df_numeric
```

```python
#discard instances with missing values
df_discard = df_numeric.dropna()
df_discard
```

```python
#treat missing values as new values
df_ternary = df_numeric.fillna(0.5)
df_ternary
```

```python
#impute missing values
df_impute = df_numeric.copy()
for col in df_numeric.columns:
    col_mode = df_impute[col].mode(dropna=True)[0]
    df_impute[col].fillna(col_mode, inplace=True)
df_impute
```

```python
#
def evaluate(clf, df):
    X = df.iloc[:, 1:]
    y = df[0]
    kf = model_selection.KFold(n_splits=5, shuffle = True, random_state = 1)
    
    f1 = []
    precision = []
    recall = []
    for train_index, test_index in kf.split(X):    
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        precision.append(metrics.precision_score(y_test, y_pred))
        recall.append(metrics.recall_score(y_test, y_pred))
        f1.append(metrics.f1_score(y_test, y_pred))
        
    print(f"    Avarage precision score: {np.mean(precision)}")
    print(f"    Avarage recall score: {np.mean(recall)}")
    print(f"    Avarage f1 score: {np.mean(f1)}")
    print()

print("Decision Tree (gini) (dicard missing values):")
evaluate(tree.DecisionTreeClassifier(), df_discard)
print("Decision Tree (entropy) (dicard missing values):")
evaluate(tree.DecisionTreeClassifier(criterion = 'entropy'), df_discard)
print("Decision Tree (gini) (ternary feature):")
evaluate(tree.DecisionTreeClassifier(), df_ternary)
print("Decision Tree (entropy) (ternary feature):")
evaluate(tree.DecisionTreeClassifier(criterion = 'entropy'), df_ternary)
print("Decision Tree (gini) (impute missing values):")
evaluate(tree.DecisionTreeClassifier(), df_impute)
print("Decision Tree (entropy) (impute missing values):")
evaluate(tree.DecisionTreeClassifier(criterion = 'entropy'), df_ternary)
print("Bernoulli Naïve Bayes (dicard missing values):")
evaluate(naive_bayes.BernoulliNB(), df_discard)
print("Multinomial Naïve Bayes (ternary feature):")
evaluate(naive_bayes.MultinomialNB(), df_ternary)
print("Categorical Naïve Bayes (ternary feature):")
evaluate(naive_bayes.CategoricalNB(), df_ternary)
print("Gaussian Naïve Bayes (ternary feature):")
evaluate(naive_bayes.GaussianNB(), df_ternary)
print("Bernoulli Naïve Bayes (impute missing values):")
evaluate(naive_bayes.BernoulliNB(), df_impute)
```
